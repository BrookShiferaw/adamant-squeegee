```python
import matplotlib.cm as cm
from numpy import *

df1 = pd.read_csv('sample.csv')
df1.insert(9,'amp','')
```

```python
tempamplitude =[]
for i in df1.amplitude:
    if i<=5:
        x = 5
    elif i>5 and i<=10:
        x = 10
    elif i>10 and i<15:
        x = 15
    elif i >15 and i<=20:
        x = 20
    elif i >20 and i<=25:
        x = 25
    else:
        x = 30
    tempamplitude.append(x)

df1.amp = tempamplitude #creates a new column to group amplitude by size
```

```python
df2 = df1.groupby(['condition', 'id','trial'])
df2 = df2.mean()
df2 = df2.reset_index()
```

```python
baseline = df2.loc[df2['condition'] == 'baseline']
baseline = baseline.reset_index(drop=True)
alcohol = df2.loc[df2['condition'] == 'alcohol']
alcohol = alcohol.reset_index(drop=True)
```

```python
baselinemicro = baseline[baseline.amplitude  < 1.5]
alcoholmicro = alcohol[alcohol.amplitude  < 1.5]

baselinemid = baseline[baseline.amplitude  <4]
alcoholmid = alcohol[alcohol.amplitude  < 4]

baselinemid = baseline[baseline.amplitude  > 1.5]
alcoholmid = alcohol[alcohol.amplitude  > 1.5] 
                     
baselinemacro = baseline[baseline.amplitude  >4]
alcoholmacro = alcohol[alcohol.amplitude  > 4]
```

**Effects of alcohol on saccade main sequence characterstics.**
- Influence of 0.05BAC on saccade peak velocity   
Alcohol decreased peak velocity [F(1, 573) = 9.03, *p* < 0.01]   
Alcohol x amplitude had greater effect on peak velocity [F(1, 570) = 13.73, *p* < 0.001]

```python
allfitb = np.polyfit(baseline.amplitude, baseline.peak_velocity, deg=1)
plt.plot(baseline.amplitude, allfitb[0] * baseline.amplitude + allfitb[1], color='green')
plt.scatter(baseline.amplitude,baseline.peak_velocity, c=baseline.peak_velocity,cmap=cm.Greens)

allfita = np.polyfit(alcohol.amplitude, alcohol.peak_velocity, deg=1)
plt.plot(alcohol.amplitude, allfita[0] * alcohol.amplitude + allfita[1], color='red')
plt.scatter(alcohol.amplitude,alcohol.peak_velocity,c=alcohol.peak_velocity,cmap=cm.Reds)
```

Differences in the effects of alcohol on small (up to 1.5deg) vs larger (2 - 12deg) saccades.

```python
microfitb = np.polyfit(baselinemicro.amplitude, baselinemicro.peak_velocity, deg=1)
microfita = np.polyfit(alcoholmicro.amplitude, alcoholmicro.peak_velocity, deg=1)
midfitb = np.polyfit(baselinemid.amplitude, baselinemid.peak_velocity, deg=1)
midfita = np.polyfit(alcoholmid.amplitude, alcoholmid.peak_velocity, deg=1)

fig2 = plt.figure(figsize=(14.0,8.0))
axes1 = fig2.add_subplot(1,2,1)
axes2 = fig2.add_subplot(1,2,2)
#axes3 = fig2.add_subplot(1,3,3)

axes1.plot(baselinemicro.amplitude, microfitb[0] * baselinemicro.amplitude + microfitb[1], color='green')
axes1.scatter(baselinemicro.amplitude,baselinemicro.peak_velocity, c=baselinemicro.peak_velocity,cmap=cm.Greens)
axes1.plot(alcoholmicro.amplitude, microfita[0] * alcoholmicro.amplitude + microfita[1], color='red')
axes1.scatter(alcoholmicro.amplitude,alcoholmicro.peak_velocity,c=alcoholmicro.peak_velocity,cmap=cm.Reds)

axes2.plot(baselinemid.amplitude, midfitb[0] * baselinemid.amplitude + midfitb[1], color='green')
axes2.scatter(baselinemid.amplitude,baselinemid.peak_velocity, c=baselinemid.peak_velocity,cmap=cm.Greens)
axes2.plot(alcoholmid.amplitude, midfita[0] * alcoholmid.amplitude + midfita[1], color='red')
axes2.scatter(alcoholmid.amplitude,alcoholmid.peak_velocity,c=alcoholmid.peak_velocity,cmap=cm.Reds)
```
